```python
def plot_df(df, title):
  df.plot(title=(title), figsize=(15,10), lw=2)
  plt.xlabel('Date')
  plt.ylabel('Power (W)')
  plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
  plt.show()
```

```python
def compute_missing_rate(df):
  missing_rate = df.isnull().sum() / len(df)
  missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'missing_rate': missing_rate})
  missing_value_df.sort_values('missing_rate', ascending = False, inplace=True)
  return missing_value_df
```

# Handling outliers


```python
def where_arbitrary_quantile(df, percent):
  high_quantile = df.quantile(percent)
  high_outliers = (df > high_quantile)
  return df.where(~high_outliers, high_quantile, axis=1)


def where_iqr(df):
  q_25 = df.quantile(0.25)
  q_75=df.quantile(0.75)
  iqr=q_75-q_25
  high_quantile = q_75+iqr*1.5
  high_outliers = (df>q_75+iqr*1.5)
  return df.where(~high_outliers, high_quantile, axis=1)
```

```python
def plot_multiple_features(df, subplots, fig_w, fig_h):
  fig, axes = plt.subplots(subplots,subplots,figsize = (fig_w, fig_h))
  for column, ax in zip(df.columns, axes.flatten()):
    df[[column]].plot(ax=ax, title=column, fontsize = 18)
    ax.tick_params(axis='x',labelsize=4)
    ax.tick_params(axis='y',labelsize=4, labelrotation=75)
    ax.xaxis.label.set_visible(False)
```

### We resample data by day for further analysis.

```python
global_consumtion_daily = global_df[['global_consumption']].resample('d').mean()
```

```python
global_consumtion_daily
```

We plot rolling mean and standard deviation to see if the data is stationary or not. Depending on the number of timestamps the rolling window has, we will obtain different results. We take rolling mean on 30 days in the example below.

```python
ma =  global_consumtion_daily.rolling(30).mean()
mstd = global_consumtion_daily.rolling(30).std()

plt.plot(global_consumtion_daily, color = 'blue', label = "Original value")
plt.plot(ma, color= 'red',  label = "Rolling mean")
plt.plot(mstd, color= 'green',  label = "Rolling std")
plt.legend()
```

```python
ma =  global_consumtion_daily.rolling(7).mean()
mstd = global_consumtion_daily.rolling(7).std()

plt.plot(global_consumtion_daily, color = 'blue', label = "Original value")
plt.plot(ma, color= 'red',  label = "Rolling mean")
plt.plot(mstd, color= 'green',  label = "Rolling std")
plt.legend()
```

We can see that if we take rolling standard deviation by month, it changes less than if take rolling standard deviation weekly.

We will decompose time series in three components : trend (responsible for the variations of mean), seasonality (responsible for the variations of standard deviation), and residual (remaning part of time series, after substracting trend and seasonality).

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
import statsmodels.api as sm
consumption_decomposed= sm.tsa.seasonal_decompose(global_consumtion_daily)
fig = consumption_decomposed.plot()
plt.show()
```

We can reuse the different components of consumption_decomposed, accessing it as shown below. 

We can clearly see from this plot of the seasonal component that we have the weekly pattern.

```python
plot_df(consumption_decomposed.seasonal['2016-12-01 10:15:00':'2017-01-01 00:00:00'], "seasonal pattern, zoom on the last month of data")
```

Because the components of time series are the series themselves, we can reconstruct some part of time series by adding up the components 

```python
plot_df((consumption_decomposed.trend+consumption_decomposed.seasonal).fillna(0), "Seasonal and trend components summed")
```

# Simplest predictive modeling

For this naive method, we make a hypothesis that the consumtpion of the next day will be equal to the mean of the previous 14 days. 

```python
rolling_mean = global_consumtion_daily["global_consumption"]['2016-02-11 10:15:00':'2017-01-01 00:00:00'].rolling(14).mean().shift(1)
plt.plot(rolling_mean, "g", label="Rolling mean")
plt.plot(global_consumtion_daily["global_consumption"]['2016-02-11 10:15:00':'2017-01-01 00:00:00'][14:], label="Actual values")
plt.legend(loc="best")
```

```python
global_consumtion_daily['prediction'] = global_consumtion_daily["global_consumption"]['2016-02-11 10:15:00':'2017-01-01 00:00:00'].rolling(14).mean().shift(1)
global_consumtion_daily[15:].head(20)
```

To estimate the goodness of fit of our model we compute several metrics on last month of data. We import the methods to compute the metrics from sklearn library and then we pass the actual values and the predictions as paramaters of these models. See the Lecture slides on the regression metrics explanation.

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
test = global_consumtion_daily['global_consumption']['2016-12-01 00:00:00':'2017-01-01 10:45:00']
prediction =  global_consumtion_daily['prediction']['2016-12-01 00:00:00':'2017-01-01 10:45:00']
print("R2 score is {}".format(r2_score(test.values, prediction.values)))
print("MSE score is {}".format(mean_squared_error(test.values, prediction.values)))
print("MAE score is {}".format(mean_absolute_error(test.values, prediction.values)))
```

Let's try another forecasting method : Exponential Smoothning. While predicting with the rolling mean, all the obervations are weithed equally. Exponential smoothing gives us the possibility to weight past observations, and changing the parameter **smoothing_level** in below function you can either increase the importance of the most recent observatons (**smoothing_level** close to 1) or the importance of the first history observations (**smoothing_level** close to 0) in your predictive model.

```python
from statsmodels.tsa.api import SimpleExpSmoothing

train, test = global_consumtion_daily["global_consumption"]['2016-02-11 10:15:00':'2016-12-01 00:00:00'].fillna(0), global_consumtion_daily["global_consumption"]['2016-12-01 00:00:00':'2017-01-01 10:45:00'].fillna(0)

rcParams['figure.figsize'] = 20, 10
# Simple Exponential Smoothing
model = SimpleExpSmoothing(train).fit(smoothing_level=0.9)
pred = model.predict(start=test.index[0], end=test.index[-1])
plt.title(model.model.params['smoothing_level'])
# plot
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(train.index, model.fittedvalues, color="red",label='Fitted values')
plt.plot(pred.index, pred, label='Simple Exponential Smoothing')
plt.legend(loc='best')
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2 score is {}".format(r2_score(test.values, pred.values)))
print("MSE score is {}".format(mean_squared_error(test.values, pred.values)))
print("MAE score is {}".format(mean_absolute_error(test.values, pred.values)))
```

We can see that this method predicts well the first timestamps of test data, but fails to take into account the seasonality of the next timestamps.

To deal with this, we will improve our previous method by adding two parameters to it :  seasonality and trend. It calls Triple Exponential Smoothing, or Holt-Winters method.

```python
from statsmodels.tsa.api import ExponentialSmoothing

train, test = global_consumtion_daily["global_consumption"]['2016-02-11 10:15:00':'2016-12-01 00:00:00'].fillna(0), global_consumtion_daily["global_consumption"]['2016-12-01 00:00:00':'2017-01-01 10:45:00'].fillna(0)
model = ExponentialSmoothing(train, seasonal='add', trend='add',seasonal_periods=7).fit(use_boxcox=True)
pred = model.predict(start=test.index[0], end=test.index[-1])

plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(train.index, model.fittedvalues, color="red", label='Fitted values')
plt.plot(pred.index, pred, label='Holt-Winters')
plt.legend(loc='best')
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2 score is {}".format(r2_score(test.values, pred.values)))
print("MSE score is {}".format(mean_squared_error(test.values, pred.values)))
print("MAE score is {}".format(mean_absolute_error(test.values, pred.values)))
```

# Feature engineering and prediction with machine learning algorithms


For more advanced technics we have to design the features. The first algorithm we try will be linear regression. Our target is the consumption we want to predict, based on the observations of a past week.

```python
dataset = global_consumtion_daily[["global_consumption"]]
dataset.columns = ["target"]
for i in range(7, 14):
    dataset["t-{}".format(i)] = dataset.target.shift(i)
dataset.head(15)
```

```python
dataset=dataset.iloc[13:]
dataset.head()
```

We design another feature : weekend. Taking advantage of the datetime format of index is very easy to do it.

```python
dataset['weekend'] = dataset.index.weekday.isin([5,6])*1
dataset.head()
```

We separate our dataset in train and test sets. Train set will be used for training a model to find a linear regression coefficients, test set will be used for validation of this model. 

```python
train, test = dataset['2016-02-11 10:15:00':'2017-01-01 00:00:00'].fillna(0), dataset['2017-01-01 00:00:00':'2017-02-09 10:45:00'].fillna(0)
```

For linear regression algorithm we have to scale the input features, to make the interpretation of coefficients easier.
Stadard scaler will transforme features such as they will have mean equal to 0 and the standard deviation equal to 1 after the transformation. 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_features_scaled = scaler.fit_transform(train.drop(['target'], axis=1))
test_features_scaled = scaler.transform(test.drop(['target'], axis=1))
```

```python
# We import Linear Regression class from sklearn library
from sklearn.linear_model import LinearRegression

# We create our model by calling the imported class constructor
modelLr = LinearRegression()

# We train our model on the train features and target, by passing them as the arguments to the model method "fit"
modelLr.fit(train_features_scaled, train['target'])

# We use trained model on the test features using the model method "predict", thus obtaining predicted target values
predictionLr = modelLr.predict(test_features_scaled)
    
# We then plot the predicted target values along with actual ones, to evaluate visually the quality of the predictive model.
plt.figure(figsize=(20, 10))
plt.plot(test['target'].values, label="real consumption")
plt.plot(predictionLr, label="prediction")
plt.legend()
```

The advantage of Linear regression is  its explainability. To find which coefficient contributes the most to the prediction, we can access the **_coef** parameter of the model, that stocks the coefficients of the regression.

We can see from the graphic below that weekend parameter plays the most important role, following by the value of consumprion the same day last week.
This kind of analysis could be very useful for domain experts, who are not only interested in the predictions, but also on the factors that influence the target variable.

```python
features_weights = pd.DataFrame(data=modelLr.coef_, index = train.drop(['target'], axis=1).columns)
features_weights.columns = ["weights"]
features_weights["absolute_value"] = features_weights["weights"].apply(np.abs)
features_weights = features_weights.sort_values(by="absolute_value", ascending=False).drop(["absolute_value"], axis=1)
    
features_weights['weights'].plot(kind='bar')
plt.hlines(y=0, xmin=0, xmax=len(features_weights))
```

# Random Forest Regressor

We use the same data, (only not scaled, because it's not needed for Random forest algorithm), for training of Random Forest regressor.
We use almost the same syntax as for linear regression : importe RandomForestRegressor, fit it on the training dataset and predict on the test features.

We pass to RandomForestRegressor the hyper-parameters of the algorithm : max depth of each decision tree in random forest, the number of trees, the max number of features to use while making a split. 

```python
from sklearn.ensemble import RandomForestRegressor
from numpy import asarray

train_features = train.drop(['target'], axis=1, inplace=False)
train_target = train['target']
test_features =  test.drop(['target'],axis=1, inplace=False)

model_RF = RandomForestRegressor(max_depth=10, random_state=0, max_features=3, n_estimators=200)
model_RF.fit(asarray(train_features), asarray(train_target))

predictionRf = model_RF.predict(test_features)
    
plt.figure(figsize=(20, 10))
plt.plot(test['target'].values, label="real consumption")
plt.plot(predictionRf, label="prediction")
plt.legend()
```

Compare now the perfomance of these two algorithmes. We can clearly see that Random Forest perfoms better, and it's not suprising because multiple regressors - decision trees, 

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2 score is {}".format(r2_score(test['target'].values, predictionRf)))
print("MSE score is {}".format(mean_squared_error(test['target'].values, predictionRf)))
print("MAE score is {}".format(mean_absolute_error(test['target'].values, predictionRf)))
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2 score is {}".format(r2_score(test['target'].values, predictionLr)))
print("MSE score is {}".format(mean_squared_error(test['target'].values, predictionLr)))
print("MAE score is {}".format(mean_absolute_error(test['target'].values, predictionLr)))
```

But is Random Forest as explainable as Linear Regression? 
We can visualize every decision tree as seen below. 

```python
import graphviz
from sklearn.tree import export_graphviz
import pydot
tree = model_RF.estimators_[9]
export_graphviz(tree, out_file = 'tree.dot', feature_names = train_features.columns, rounded = True, precision = 1)
(graph, ) = pydot.graph_from_dot_file('tree.dot')
graphviz.Source(graph)
```

```python
small_model_RF = RandomForestRegressor(max_depth=3, random_state=0, max_features=3, n_estimators=10)
small_model_RF.fit(asarray(train_features), asarray(train_target))

predictions_small_rf = small_model_RF.predict(test_features)
    
plt.figure(figsize=(20, 10))
plt.plot(test['target'].values, label="real consumption")
plt.plot(predictions_small_rf, label="prediction")
plt.legend()
```

```python
tree = small_model_RF.estimators_[2]
export_graphviz(tree, out_file = 'tree.dot', feature_names = train_features.columns, rounded = True, precision = 1)
(graph, ) = pydot.graph_from_dot_file('tree.dot')
graphviz.Source(graph)
```

We can also access the importance of every variable in the model. Like coef_ parameter in linear regression model, random forest model has feature_importances_ parameter, that we can plot to visualize the weight of every variable to explaing our target.

```python
features_weights = pd.DataFrame(data=model_RF.feature_importances_, index = train_features.columns)
features_weights.columns = ["feature_importances"]
features_weights["absolute_value"] = features_weights["feature_importances"].apply(np.abs)
features_weights = features_weights.sort_values(by="absolute_value", ascending=False).drop(["absolute_value"], axis=1)
    
features_weights['feature_importances'].plot(kind='bar')
plt.hlines(y=0, xmin=0, xmax=len(features_weights))
```
