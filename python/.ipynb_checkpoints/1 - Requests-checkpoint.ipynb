## Exercice : Getting data

Using the response from 'https://api.github.com/events' create an unique list of login associated to an event.

```python
token = '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'
headers = {"Authorization": f"token {token}"}
response = requests.get('https://api.github.com/events', headers=headers)
```

```python
## exercice

events = response.json()

les_logins = []

for event in events:
    les_logins.append(event['actor']['login'])
print (les_logins)
```

```python
events = response.json()
logins=[event["actor"]["login"] for event in events]

print(logins)
print("----------")
for index, login in enumerate(logins) :
    if "[bot]" in login:
        logins.pop(index)
    
logins = list(set(logins))
print(logins)
```

## Exercice : Filter data

Using the list of user previously created, get the profil url associate to each login with the highest number of followers.

https://docs.github.com/en/rest/reference/search#search-users

```python
## exercice




les_profils = []
token = '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'
headers = {"Authorization": f"token {token}"}



for login in logins[:10]:

    response = requests.get("https://api.github.com/search/users?q="+login, headers=headers).json()
    if(response['total_count'] >0):
        #les_profils.append([login, response['login'], response['followers']])
        #print(response)
        #print(response['items'][0]['url'])
        #print('--------------')
        followers = requests.get(response['items'][0]['url']+"/followers", headers=headers).json()
        #print(len(followers))
        les_profils.append([login, response['items'][0]['url'], len(followers)])

for profil in les_profils:
    print(profil)

    
#print(les_profils)


```

## Exercice : Use data within a web api

For the 2 Top folowers profils, use the hideuri api to shorten their url

https://hideuri.com/docs


```python
## exercice

from operator import itemgetter

top = sorted(les_profils, key=itemgetter(2), reverse=True)
top = top[:2]
print(top)
print("--------------")

for i in range(len(top)):
    url = top[i][1]
    params = {"url": str(url)}

    response = requests.post('https://hideuri.com/api/v1/shorten', data=params).json()
    top[i].append(response['result_url'])
    print(top[i])
    
```
